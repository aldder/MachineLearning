```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    """
    Frame a time series as a supervised learning dataset.
    Arguments:
        data: Sequence of observations as a list or NumPy array.
        n_in: Number of lag observations as input (X).
        n_out: Number of observations as output (y).
        dropnan: Boolean whether or not to drop rows with NaN values.
    Returns:
        Pandas DataFrame of series framed for supervised learning.
    """
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
    # forecast sequence (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
    # put it all together
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    # drop rows with NaN values
    if dropnan:
        agg.dropna(inplace=True)
    return agg

data = series_to_supervised(data, n_in=30, n_out=7, dropnan=True)
data
```

```python
for n in [7,14,21,28]:
    data[f'Avg-{n}'] = data[[f'var1(t-{i+1})' for i in range(n)]].mean(axis=1)
    
for n in [7,14,21,28]:
    data[f'Med-{n}'] = data[[f'var1(t-{i+1})' for i in range(n)]].median(axis=1)
    
data
```

```python
data['weekday'] = data.index.weekday_name
data['month'] = data.index.month_name()
data['quarter'] = data.index.quarter
data['doy'] = data.index.dayofyear
data['dim'] = data.index.daysinmonth

data = data.merge(pd.get_dummies(data['weekday'], drop_first=True), 
                              left_index=True, right_index=True, how='inner').drop(['weekday'],1)
data = data.merge(pd.get_dummies(data['month'], drop_first=True), 
                              left_index=True, right_index=True, how='inner').drop(['month'],1)
data = data.merge(pd.get_dummies(data['quarter'], drop_first=True, prefix='QRT'),
                              left_index=True, right_index=True, how='inner').drop(['quarter'],1)

data
```

```python
from keras.models import Sequential
from keras.layers import Dense
import keras.backend as K
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from bokeh.models import ColumnDataSource, HoverTool, RangeTool
from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import gridplot, column
output_notebook()

def build_model(X_train, y_train, X_test, y_test, params):
    
    earlyStopping = EarlyStopping(patience=params['earlyStopping'])
    checkPoint = ModelCheckpoint(f'best_model.h5', monitor='val_loss', verbose=0, 
                                 save_best_only=True, save_weights_only=True, mode='auto', period=1)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=params['reduce_factor'], patience=params['reduce_patience'],
                                  verbose=0, mode='auto', min_delta=0.0001, cooldown=0, min_lr=0)

    K.clear_session()
    model = Sequential()
    model.add(Dense(params['n_hidden_neurons'], activation=params['activation'], input_dim=X_train.shape[1]))
    model.add(Dense(7))
    model.compile(loss='mse', optimizer='adam', metrics=['mae'])
    print(model.summary())
    
    # fit network
    history = model.fit(X_train, y_train, epochs=params['epochs'], batch_size=params['batch_size'], verbose=0, 
                        validation_split=0.2, callbacks=[reduce_lr, earlyStopping, checkPoint])
    
    model.load_weights(f'best_model.h5')
    model.compile(loss='mse', optimizer='adam', metrics=['mae'])
    
    print('Evaluation', model.evaluate(X_test, y_test, verbose=1))

    p = figure(plot_width=400, plot_height=300, tools=['pan','wheel_zoom','box_zoom','reset','save'],
              toolbar_location='above', title='Train/Test loss', x_axis_label='epoch', y_axis_label='loss')
    p.line(history.epoch, history.history['loss'], legend='Train loss', line_width=2)
    p.line(history.epoch, history.history['val_loss'], color='orangered', legend='Test loss', line_width=2)
    q = figure(plot_width=400, plot_height=300, tools=['pan','wheel_zoom','box_zoom','reset','save'],
                  toolbar_location='above', title='Train/Test mae', x_axis_label='epoch', y_axis_label='mae')
    q.line(history.epoch, history.history['mean_absolute_error'], legend='Train MAE', line_width=2)
    q.line(history.epoch, history.history['val_mean_absolute_error'], color='orangered', legend='Test MAE', line_width=2)
    r = gridplot([[p,q]])
    show(r)
            
    return model, X_train, y_train, X_test, y_test
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler

X = data[features].values
y = data[y].values

X_scaler = MinMaxScaler(feature_range=(0,1))
y_scaler = MinMaxScaler(feature_range=(0,1))

X = X_scaler.fit_transform(X)
y = y_scaler.fit_transform(y)

test_size = 0.2

print('full dataset size', len(X))
print('testset size', int(len(X) * test_size))

X_train = X[:-(int(len(X) * test_size))] 
X_test = X[-(int(len(X) * test_size)):]

print(X_train.shape, X_test.shape)

y_train = y[:-(int(len(y) * test_size))]
y_test = y[-(int(len(y) * test_size)):]

print(y_train.shape, y_test.shape)
```

```python
params = {
    'n_hidden_neurons': 100,
    'activation': 'linear',
    'epochs': 1000,
    'batch_size': 100,
    'earlyStopping': 50,
    'reduce_factor': 0.1,
    'reduce_patience': 10,
}

model, X_train, y_train, X_test, y_test = build_model(X_train, y_train, X_test, y_test, params)
```

```python
y_pred = y_scaler.inverse_transform(model.predict(X_test))
y_test = y_scaler.inverse_transform(y_test)
```

```python
from sklearn import metrics

r2s = []
rmses = []
maes = []
for i in range(7):
    
    r2 = metrics.r2_score(y_test[:,i], y_pred[:,i])
    rmse = np.sqrt(metrics.mean_squared_error(y_test[:,i], y_pred[:,i]))
    mae = metrics.mean_absolute_error(y_test[:,i], y_pred[:,i])
    
    print(f'\neval d+{i+1}')
    print(f'r2: {r2}')
    print(f'rmse: {rmse}')
    print(f'mae: {mae}')
    
    r2s.append(r2)
    rmses.append(rmse)
    maes.append(mae)
print()   
print(np.mean(r2s))
print(np.mean(rmses))
print(np.mean(maes))
```
