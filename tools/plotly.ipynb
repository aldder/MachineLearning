```python
import numpy as np
import pandas as pd

from plotly import subplots
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)

import quandl
```

```python
ECR = ['EUR/USD','EUR/JPY','EUR/CAD','EUR/GBP','EUR/CNY','EUR/AUD','EUR/RUB']

exchange_rates = pd.concat([
    quandl.get('ECB/EURUSD', start_date='2019-01-01'),
    quandl.get('ECB/EURJPY', start_date='2019-01-01'),
    quandl.get('ECB/EURCAD', start_date='2019-01-01'),
    quandl.get('ECB/EURGBP', start_date='2019-01-01'),
    quandl.get('ECB/EURCNY', start_date='2019-01-01'),
    quandl.get('ECB/EURAUD', start_date='2019-01-01'),
    quandl.get('ECB/EURRUB', start_date='2019-01-01')
], 1)
exchange_rates.columns = ECR
exchange_rates
```

```python
g1 = go.Scatter(x=exchange_rates.index, y=exchange_rates['EUR/USD'], name='EUR/USD')
g2 = go.Scatter(x=exchange_rates.index, y=exchange_rates['EUR/JPY'], name='EUR/JPY')
g3 = go.Scatter(x=exchange_rates.index, y=exchange_rates['EUR/CAD'], name='EUR/CAD')
g4 = go.Scatter(x=exchange_rates.index, y=exchange_rates['EUR/GBP'], name='EUR/GBP')
g5 = go.Scatter(x=exchange_rates.index, y=exchange_rates['EUR/CNY'], name='EUR/CNY')
g6 = go.Scatter(x=exchange_rates.index, y=exchange_rates['EUR/AUD'], name='EUR/AUD')
g7 = go.Scatter(x=exchange_rates.index, y=exchange_rates['EUR/RUB'], name='EUR/RUB')

fig = subplots.make_subplots(rows=4, cols=2, specs=[[{'colspan':2}, None], [{}, {}], [{}, {}], [{}, {}]],
                             subplot_titles=ECR, print_grid=False)
fig.update_layout(height=800, title_text="EUR exchange rates", showlegend=False)

fig.append_trace(g1, 1, 1)
fig.append_trace(g2, 2, 1)
fig.append_trace(g3, 2, 2)
fig.append_trace(g4, 3, 1)
fig.append_trace(g5, 3, 2)
fig.append_trace(g6, 4, 1)
fig.append_trace(g7, 4, 2)

iplot(fig)
```
